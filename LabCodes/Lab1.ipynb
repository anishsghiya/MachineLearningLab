```python
file = open('/content/test.txt', "r")
freq = {}
for line in file :
    # print(line) 
    words = line.split(" ")
    for word in words :
        # print(word)
        if word in freq : 
            freq[word]+=1
        else : 
            freq[word] = 1
file.close()
```
